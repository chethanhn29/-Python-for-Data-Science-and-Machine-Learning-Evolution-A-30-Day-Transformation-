## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt

# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Plot a bar plot of the average sepal length by species
plt.bar(x=iris['species'], height=iris['sepal_length'])
plt.show()

# Plot a scatter plot of sepal width vs sepal length, colored by species
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
plt.scatter(x=iris['sepal_width'], y=iris['sepal_length'], c=iris['species'].apply(lambda x: colors[x]))
plt.show()

# Plot a histogram of the sepal length values, with 15 bins
plt.hist(x=iris['sepal_length'], bins=15)
plt.show()

# Plot a box plot of sepal length by species
plt.boxplot(x=iris['sepal_length'], by=iris['species'])
plt.show()

# Plot a kernel density estimate of the sepal length values
from scipy.stats import gaussian_kde
density = gaussian_kde(iris['sepal_length'])
xs = np.linspace(iris['sepal_length'].min(), iris['sepal_length'].max(), 100)
plt.plot(xs, density(xs))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**3

# Basic line plot
plt.plot(x, y1)
plt.show()

# Line plot with multiple lines
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.legend()
plt.show()

# Line plot with custom colors and line styles
plt.plot(x, y1, 'r--', label='y1')
plt.plot(x, y2, 'b-', label='y2')
plt.legend()
plt.show()

# Scatter plot
plt.scatter(x, y1)
plt.show()

# Scatter plot with different marker styles and sizes
plt.scatter(x, y1, marker='x', s=100)
plt.scatter(x, y2, marker='o', s=50)
plt.show()


```

```python
# Bar plot
plt.bar(x, y1)
plt.show()

# Horizontal bar plot
plt.barh(x, y1)
plt.show()

# Histogram
plt.hist(y1, bins=10)
plt.show()

# Pie chart
labels = ['apple', 'banana', 'orange', 'grape']
sizes = [20, 30, 40, 10]
plt.pie(sizes, labels=labels)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

# Seaborn
it is a data visualization library built on top of Matplotlib that provides a higher-level interface for creating a variety of plots and charts. It is particularly useful for statistical data visualization and has several built-in functions for plotting statistical data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Plot a bar plot of the average sepal length by species
sns.barplot(x='species', y='sepal_length', data=iris)
plt.show()

# Plot a scatter plot of sepal width vs sepal length, colored by species
sns.scatterplot(x='sepal_width', y='sepal_length', hue='species', data=iris)
plt.show()

# Plot a histogram of the sepal length values, with 15 bins
sns.histplot(x='sepal_length', data=iris, bins=15)
plt.show()

# Plot a box plot of sepal length by species
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()

# Plot a kernel density estimate of the sepal length values
sns.kdeplot(iris['sepal_length'], shade=True)
plt.show()
```

```python
import seaborn as sns

# Load a sample dataset
tips = sns.load_dataset('tips')

# Plot a line plot
sns.lineplot(x='total_bill', y='tip', data=tips)

# Show the plot
plt.show()
```

This will create a line plot with the tip values on the y-axis and the corresponding total_bill values on the x-axis, using the tips dataset from Seaborn's built-in datasets.

To customize the appearance of the plot, you can use various parameters of the lineplot function, such as color, linewidth, and style. You can also use the hue parameter to color the lines by a categorical variable.

Here is an example of how to use the hue parameter and customize the appearance of the plot:

```python
import seaborn as sns

# Load a sample dataset
tips = sns.load_dataset('tips')

# Plot a line plot with different colors for each category
sns.lineplot(x='total_bill', y='tip', hue='day', data=tips,
             color='r', linewidth=2.5, style='--')

# Show the plot
plt.show()
```
